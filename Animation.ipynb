### Animation with matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation,FFMpegFileWriter
import matplotlib.animation as animation
```

```python
# Create figure object
fig, ax = plt.subplots()

#empty data
xdata, ydata = [], [] # actual data to be filled in later

#plotting object
ln, = plt.plot([], [],  animated=True);



#axes limit
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-1.1, 1.1)


ln.set_data(xdata,ydata)
```

```python
# a range from -pi to +pi , later on to be used as number for animation frame
f = np.linspace(-np.pi,np.pi, 200) 
# there are 200 points from -pi to pi
f
```

```python
def init():
    ln.set_data([], [])
    return ln,
```

```python
def update(i):
    x = np.linspace(-np.pi,np.pi, 200)
    y= np.sin(2 * np.pi * (x - 0.01 * i))
    ln.set_data(x, y)
   
    
    return ln,
```

```python
ani = animation.FuncAnimation(fig, update, init_func=init,frames=f,
                     blit=True, interval = 2.5)
```

```python

FFwriter =FFMpegFileWriter()
mywriter=FFMpegFileWriter( fps=25, extra_args=['-vcodec', 'libx264'])
ani.save('D:/rani/sin_animate.mp4', writer=mywriter)
```

```python
plt.close()
```

```python

import pycountry
import pandas as pd
URL_DATASET = r'https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv'
df = pd.read_csv(URL_DATASET)
print(df.head(3))  # Get first 3 entries in the dataframe
```

```python
print(df.tail(3))  # Get last 3 entries in the dataframe
```

```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from time import sleep


```

```python

####  Create list of all dates
list_dates = df['Date'].unique()
```

```python
print(list_dates) 


```

```python
## Also create ax object
fig, ax = plt.subplots(figsize=(15, 8))
```

```python
# We will animate for these 5 countries only
list_countries = ['India', 'China', 'US', 'Italy', 'Spain']
```

```python

# colors for the 5 horizontal bars
list_colors = ['black', 'red', 'green', 'blue', 'yellow']
```

```python
# plot_bar() is the call back function used in FuncAnimation class object
def plot_bar(some_date):
    df2 = df[df['Date'].eq(some_date)]
    ax.clear()
    # Only take Confirmed column in descending order
    df3 = df2.sort_values(by = 'Confirmed', ascending = False)
    
    # Select the top 5 Confirmed countries
    df4 = df3[df3['Country'].isin(list_countries)]

    sleep(0.2)  # To slow down the animation
    # ax.barh() makes a vertical bar plot.
    return ax.barh(df4['Country'], df4['Confirmed'], color= list_colors)

```

```python
covid_anim = animation.FuncAnimation(fig = fig, func = plot_bar,
                    frames= list_dates, blit=True,
                    interval=20)
```

```python

FFwriter =FFMpegFileWriter()
mywriter=FFMpegFileWriter( fps=25, extra_args=['-vcodec', 'libx264'])
covid_anim.save('D:/rani/covid_animate.mp4', writer=mywriter)
```
